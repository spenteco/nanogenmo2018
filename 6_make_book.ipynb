### Shovel the stories through Latex

```python
import codecs, re, glob

stories = {}

for p in glob.glob('cleaned_stories/*.txt'):
    
    n = int(p.split('/')[-1].replace('.txt', ''))
    
    stories[n] = codecs.open(p, 'r', encoding='utf-8').read()
    
print 'len(stories)', len(stories)
```

```python
from mako.template import Template
    
document_template = Template(filename='to_pdf/templates/document.tex')
blank_page_template = Template(filename='to_pdf/templates/blank_page.tex')
title_page_template = Template(filename='to_pdf/templates/title_page.tex')
copyright_template = Template(filename='to_pdf/templates/copyright.tex')
header_and_footer_template = Template(filename='to_pdf/templates/header_and_footer.tex')

# ----------------------------------------------------------------------

tex_body = [r'\newpage', r'\pagenumbering{arabic}', r'\pagestyle{fancy}', r'\fancyhf{}',
            
            r'\fancyhead[RE]{\textsc{Velvet Black Sky}}',
            r'\fancyhead[LO]{\textsc{Stephen Pentecost}}',
            r'\fancyhead[LE,RO]{\thepage}'
            
            #r'\fancyhead[LE,RO]{Velvet Black Sky}',
            #r'\fancyhead[RE,LO]{\thepage}'
           ]

for story_n in sorted(stories.keys()):
             
    tex_body.append(r'\newpage')
    tex_body.append(r'\begin{center}')
    tex_body.append(r'\textbf{' + str(story_n + 1) + '}')
    tex_body.append(r'\end{center}')
    
    tex_body.append(stories[story_n])
    
tex_body.append(r'\newpage')

# ----------------------------------------------------------------------

tex_pages = []
tex_pages.append(blank_page_template.render())

tex_pages.append(title_page_template.render(vspace_setting=20.0, 
                                            title='Velvet Black Sky', 
                                            author='Stephen Pentecost'))
                                            
tex_pages.append(copyright_template.render())
                                            
tex_pages.append(blank_page_template.render())
tex_pages.append(blank_page_template.render())

tex_pages.append('\n'.join(tex_body).replace('$', r'\$').replace('&', r'\&'))

final_document = document_template.render(tex_pages=tex_pages)

f = codecs.open('to_pdf/velvet_black_sky.tex', 'w', encoding='utf-8')
f.write(final_document)
f.close()

print 'Done!'
```
